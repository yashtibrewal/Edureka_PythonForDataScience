### Answer 1

```python
num = int(input(""))
for i in range(1,num):
    if num % i ==0:
        if i % 2 ==0:
            print(i," even")
        else:
            print(i," odd") 
```

### Answer 4

```python
sentence = input()
letters = 0
digits = 0
for char in sentence:
    if char.isdigit():
        digits += 1
    if char.isalpha():
        letters += 1
print("LETTERS ",letters)
print("DIGITS ",digits)
```
